## SECTION B : FILTERING AND CONDITIONAL SELECTION

```python
# selecting rows of students who completed the test preparation course
student_performance[student_performance["test preparation course"] == "completed"]
```

```python
# Students who scored above 70 in math. i.e displaying the rows with score in math greater than 70.
student_performance[student_performance["math score"] > 70 ]
```

```python
# Female students who scored at least 65 in all three subjects
student_performance[(student_performance["gender"] == "female") & (student_performance["math score"] >= 65) &
 (student_performance["reading score"] >= 65) &
  (student_performance["writing score"] >= 65)]
```

```python
# Number of Students Whose Lunch are Free/Reduced
lunch_count = student_performance["lunch"].value_counts()
student_free_reduced_lunch ="Number of Student free/reduced lunch:", lunch_count["free/reduced"]
student_free_reduced_lunch
```

## SECTION C: GROUPBY & AGGREGATION


```python
# Grouping Student Average Math Scores By Gender
average_mathscores = student_performance.groupby('gender')['math score'].mean()
average_mathscores
```

```python
# Grouping student average scores(math,reading,writing) by parental level of education
average_score_numerical = student_performance.groupby('parental level of education')[['math score','reading score','writing score']].mean()
average_score_numerical
```

```python
## Comparing average scores of students with vs without test preparation (note:'none'refers to students who did notc omplete the test preparation course).
avg_scoreswith_withoutTP = student_performance.groupby('test preparation course')[['math score','reading score','writing score']].mean()
avg_scoreswith_withoutTP
```

 Students with test preparation course have higher marks than the ones without, perhaps, the test preparation have significant impact on overall student performance.

## SECTION E: INTERPRETATION


Does test preparation appear to improve performance?

Students who completed the test preparation course achieved higher average scores in mathematics, reading, and writing than those who did not complete the course. This suggests that the test preparation program has a positive impact on academic performance. The improved scores indicate that students may benefit from the skills and strategies taught during the prep course. As a result, participation in the test preparation course appears to enhance overall student achievement. Therefore, the test preparation course appear to improve performance.
